```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import AgglomerativeClustering
from pandas.core.common import random_state
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')
```

Hierarchical Clustering

```python
df= pd.read_csv('D:/Assignments/crime_data.csv')
df
```

```python
df.info()
```

```python
# Normalization function 
scaling=MinMaxScaler()
df_norm=scaling.fit_transform(df.iloc[:,1:])
```

```python
df_norm
```

```python
# create dendrogram

from scipy.cluster.hierarchy import linkage

Z = linkage(df_norm, method='average', metric= 'euclidean')
plt.figure(figsize=(15,5))
plt.title("Dendrogram")
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(Z, leaf_rotation=0, leaf_font_size=8)
plt.show()
```

```python
# create clusters
hc=AgglomerativeClustering(n_clusters=4,affinity="euclidean",linkage='complete')
hc
```

```python
#Build cluster algorithm
y_huc=hc.fit_predict(df_norm)
cluster=pd.DataFrame(y_huc,columns=['cluster'])
```

```python
df['Cluster']=hc.labels_
```

```python
df.head(15)
```

```python
df.iloc[:,1:].groupby(df.Cluster).mean()
```

From above information , we can say that crime in cluster '0' and '3' are more

K-Means Clustering

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
crime_kmeans=pd.read_csv("D:/Assignments/crime_data.csv")
```

```python
crime_kmeans.head()
```

```python
scaler=StandardScaler()
crime_std=scaler.fit_transform(crime_kmeans.iloc[:,1:])
crime_std
```

```python
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,random_state=0)
  kmeans.fit(crime_std)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss) 
plt.title("Elbow Method")
plt.xlabel('Number of clusters') 
plt.ylabel('wcss')
plt.show()
```

Cluster algorithm K=4

```python
wcss
```

```python
clusters_4 = KMeans(n_clusters=4, random_state= 0)
clusters_4.fit(crime_std)
```

```python
clusters_4.labels_
```

```python
#Assign clusters to the data set
crime_kmeans['clusterid_new'] = clusters_4.labels_
```

```python
crime_kmeans.head()
```

```python
crime_kmeans.groupby('clusterid_new').mean()
```

From above information . we can say that crime in cluster '1' and '3' are more

Cluster algorithm K=5
